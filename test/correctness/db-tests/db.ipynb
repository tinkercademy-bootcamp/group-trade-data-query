```python
import pandas as pd
```

```python
db = pd.read_csv('../../../data/raw/trades-example.csv')
```

```python
class Query:
    symbol_id = None
    start_time_point = None
    end_time_point = None
    resolution = 0
    metrics = 0

    def __init__(self, symbol_id, start_time_point, end_time_point, resolution, metrics): 
        self.symbol_id = symbol_id
        self.start_time_point = start_time_point
        self.end_time_point = end_time_point
        self.resolution = resolution
        self.metrics = metrics
```

```python
def get_data(q: Query):
    temp = db[(db['created_at'] >= q.start_time_point) & (db['created_at'] <= q.end_time_point) & (db['symbol_id'] == q.symbol_id)]
    bins = list(range(q.start_time_point, q.end_time_point, q.resolution))
    ans = {}
    
    if q.metrics & 1:
        ans['max_price'] = temp.groupby(pd.cut(temp['created_at'], bins))['price'].max()
        ans['min_price'] = temp.groupby(pd.cut(temp['created_at'], bins))['price'].min()
    return ans
```

```python
q = Query(
    symbol_id=1,
    start_time_point=1747267200212439705,
    end_time_point=1747267200274719691,
    resolution=10000000,
    metrics=1 # Example metrics for max and min price
)
```

```python
get_data(q)
```
